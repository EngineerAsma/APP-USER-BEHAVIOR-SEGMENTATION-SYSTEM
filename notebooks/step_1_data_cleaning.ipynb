```python
# step  1 : data collection
import pandas as pd

data = pd.read_csv("C:\\Users\\User\\.vscode\\APP USER BEHAVIOR SEGMENTATION SYSTEM\\data\\app_user_behavior_dataset.csv")

df = pd.DataFrame(data)
```

```python
# step 2 : data understanding
print(df.shape)
print(df.columns)
print(df.info())
print(df.head())
print(df.describe())
num_cols =  df.select_dtypes("number").columns
print(num_cols)

cat_cols = df.select_dtypes("object").columns
print(cat_cols)
```

```python
#  data cleaning
# handling missing values
print(df.isnull().sum())
median_val = df["rating_given"].median()
df["rating_given"].fillna(median_val, inplace=True)
print(df.isnull().sum())
print(df.duplicated())
print(df.nunique())
```

```python
# correcting inconsistent categroical labels
for col in cat_cols:
    df[col] = df[col].astype(str).str.lower().str.strip()

# ensuring consistency among numerical columns
print(df[num_cols].min()) # no negative values
```

```python
df.to_csv("C:\\Users\\User\\.vscode\\APP USER BEHAVIOR SEGMENTATION SYSTEM\\data\\cleaned_data.csv", index=False)
```
